```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=divvy_2019, x='usertype', y='tripduration')

plt.title('Trip Duration by Usertype')
plt.xlabel('Usertype')
plt.ylabel('Trip Duration (seconds)')
plt.ylim(0, 3000)  # zoom in to cut extreme outliers
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
age_gender = divvy_2019.groupby(['age_group', 'gender']).size().unstack()
age_gender.plot(kind='bar', stacked=True, figsize=(10,5))
plt.title('Age Group Distribution by Gender')
plt.ylabel('Trip Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
top_routes = divvy_2019['route'].value_counts().head(10)
top_routes.plot(kind='barh', figsize=(10,5))
plt.title('Top 10 Most Frequent Routes')
plt.xlabel('Trip Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
